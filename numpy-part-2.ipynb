**2-D array**
For multipying two matrices the required condition is,
**no.of columns of first_matrix=no.of rows of second_matrix**

```python
p=np.array([[1,2,3],
            [4,5,6]])
q=np.array([[[1,3,5],
             [4,7,9],
             [5,3,8]]])
p.dot(q)  # q.dot(p) will not work since no.of columns of q and no.of rows of p are not equal
```
