**concatenate 2 strings index wise**

```python
first_name=["pavan", "varun", "allu"]
last_name=["kalyan", "tej", "arjun"]
l=len(first_name)
name=[]
for i in range(l):
    x=first_name[i] +""+ last_name[i]
    name.append(x)
print(name)
```

**how many times a particular  item is occuring in a list**

```python
l=[1,2,4,2,3]
num=int(input("Enter the number"))
count=0
for i in l:
    if num==i:
        count=count+1
print(num,"is occuring",count,"times")
```

**remove duplicate value from the list**

```python
l=[1,2,3,3,4,4,5,6]
for i in l:
    count=0
    for j in l:
        if i==j:
            count=count+1
            if count>1:
                l.remove(j)
print(l)
```
