**map functions **

```python
def attendence(attendence):
    if attendence>80:
        return "Good"
    elif attendence>60:
        return "ok"
    else:
        return "Need to improve"

attendence(61)
```

**what if we have a list-we can use a loop right?**

```python
l=[23,65,78,89,54]
for i in l:
    print(attendence(i))
```

```python
list(map(attendence,l))   #using map() function
```

```python
set(map(attendence,l))   #using map() function
```

**Filter Functions**

```python
def check(a):
    vowels=['a','e','i','o','u']
    if a in vowels:
        return True
    else:
        return False
```

```python
check('a')
```

```python
l=['a','b','e','i']
m=list(map(check,l))
m
```

```python
f=list(filter(check,l))
f
```
