**write a python script to generate and print a dictionary that contains a number(b/n 1 & n)in the form (x:x*x)[1:1,2;4,3:9.....]**

```python
n=int(input("Enter the number"))
d={}
for i in range(1,n+1):
    d[i]=i*i
print(d)
```

**write a program to sort the dictionary by keys**

```python
d2={
    "a":23,
    "c":25,
    "b":22
}

s=sorted(d2)
d3={}
for i in s:
    d3[i]=d2[i]
print(d3)
```

**write a python program to get the top three cities by temperature which are stored in a dictionary**

```python
temp={
    'delhi':45,
    'mumbai':26,
    'bangalore':18,
    'chennai':35,
    'lucknow':40
}
val=sorted(temp.values(),reverse=True)
print(val)
for i in range(0,3): #this will iterate through sorted list
    for j in temp:
        if val[i]==temp[j]:
            print(j)
    
    
```

**check how many times a word is there in a sentence**

```python
a="fear leads to anger and anger leads to hate "
b=a.split()
d={}
for i in b:
   
    if d.get(i)==None:  # or can use  "if i not in d:"" 
    
        d[i]=1
    else:
        d[i]=d[i]+1
print(d)
        
```
