```python
import pandas as pd 
```

```python
d={
    'country':['India','China','Japan','Bangladesh','Singapur'],
    'Gold':[12,30,11,3,None]
}
a=pd.DataFrame(d)
a
```

```python
a['Rank']=a['Gold'].rank(ascending=False)
a.sort_values('Rank',inplace=True)
a
```

```python
a['Rank']=a['Gold'].rank(ascending=False,na_option='bottom')
a.sort_values('Rank',inplace=True)
a
```

**what if data is same**

```python
d={
    'country':['India','China','Japan','Bangladesh','Singapur'],
    'Gold':[12,30,12,3,None]
}
a=pd.DataFrame(d)
a
```

```python
a['Rank']=a['Gold'].rank(ascending=False)
a.sort_values('Rank')
a
```

```python
a['Rank']=a['Gold'].rank(ascending=False,method='min')
a.sort_values('Rank')
a
```
