**Control Statements****BREAK STATEMENT**

```python
i=0
while True:
    print(i)
    i=i+1
    if i==5:
        break
    
```

```python
count = 0
num = 1
while count<21:
    if num % 2 != 0:
        count += 1
    if count ==20:
        print(num)
        break
    num +=1
```

```python
for i in range(1,31):
    if i % 2 != 0:
        print(i)
```

**continue statement**

```python
i=0
while i<=5:
    i=i+1
    if i==3:
        continue
    print(i)
```

**nested loops**

```python
i=1
while i<=4:
    j=1
    while j<=3:
        print(j, end =" ")
        j=j+1
    i=i+1
    print("\n")
```

printting pattern

```python
i=1
while i<=5:
    j=1
    while j<=i:
        print("1", end = " ")
        j=j+1
    i=i+2
   
    print("\n")
```

prime numbers between  2-20 

```python
i=2
while i<=20:
    j=2
    while j<=i//2+1:
        if i==2:
            print(i)
        if i % j ==0:
            break
        if j==i//2+1:
            print(i)
        j=j+1
    i=i+1
```
