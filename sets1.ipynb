***check whether the given string is a palangram or not ***palangran is a sentence that has every letter in english alphabet(a-z)eg:"The quick brown  fox jumps over the lazy dog"

```python
a = "The quick brown fox jumps over the lazy dog"
a = a.lower()
b = set(a)
count = 0
for i in b:
    if i.isalpha() == True:
        count = count + 1
if count == 26:
    print("This is a palangram")
else:
    print("This is not a palangram")
```
