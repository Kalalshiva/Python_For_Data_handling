**Comprehension- is an elegant way to crate list,dictionar,sets in concise manner**

```python
l=[]
for i in range(1,6):
    l.append(i**2)
print(l)
```

```python
l1=[]
for i in range(1,12):
    if i%2==0:
        l1.append(i)
print(l1)
```

```python
l1=[i for i in range(1,12) if i%2==0]   #list comprehension
l1
```

```python
d1={}
for i in range(1,6):
    d1[i]=i**2
    
print(d1)
```

```python
d2={i:i**2 for i in range(1,6) }  #dictionary comprehension
d2
```
