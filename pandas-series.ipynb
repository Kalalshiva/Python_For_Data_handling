```python
import pandas as pd
```

**creating Series**

```python
l1=["pavan","kalyan","tej","aman"]
```

```python
s=pd.Series(l1)
s
```

index is added to series data by default.data type of string but in pandas it is represented as object

```python
l2=[True,False,True]
s1=pd.Series(l2)
s1
```

**Creating Series from dictionary**

```python
d={
    'Bengaluru':123,
    'Delhi':456,
    'Mumbai':567
}
```

```python
s2=pd.Series(d)
s2
```

**Custommized index**

```python
l=[23,45,66]
k=pd.Series(l)
k
```

```python
l=[23,45,66]
l4=['A','B','C']
k=pd.Series(l,index=l4)
k
```
