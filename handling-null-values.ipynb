```python
import pandas as pd
```

```python
d={
    'students':['shiva','mahi','pavan','kalyan','None'],
    'marks':[None,14,23,15,None],
    'age':[18,18,None,None,None]
}
    
d
```

```python
df=pd.DataFrame(d)
df
```

```python
df.dropna()
```

```python
df.dropna(how='all')
df
```

```python
df.dropna(how='all',inplace=True)
df
```

```python
df['age'] = df['age'].fillna(18)
df
```
